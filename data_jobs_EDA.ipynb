Q2: Data Scientist experience level pay distribution?

```python
d_science = ds_23[ds_23["job_title"]=="Data Scientist"]
d_science
```

```python
exp = d_science["experience_level"]
```

```python
import matplotlib.ticker as ticker
```

```python
colors = []
for i in exp:
    if i=="Senior":
        colors.append("#cc2900")
    else:
        colors.append("#ff9980")
plt.bar(d_science["experience_level"],d_science["salary_in_usd"],color=colors)
#for editing y-ticks
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: '${:,.0f}'.format(x)))
plt.title("Pay distribution according to experience level")
plt.ylabel("Salary in USD$     ")
plt.xlabel("Experience Level")
plt.savefig("Pay distribution.jpeg")
plt.show()
```
